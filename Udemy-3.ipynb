## 9. Looping Through Dictionaries


```python
athlete = {
    "Name" : "Aldrin",
    "Age" : "Twenty One",
    "Height" : "Five Feet 10 Inches",
    "Weight" : "Sixty Kilos"
}
```

```python
# Looping through only the keys 

for keys in athlete.keys():      # in For you can write anything instead of keys,but that should be repeated in print().
    print(keys)
```

```python
athlete.keys()
```

```python
# Looping through only the values 

for values in athlete.values():
    print(values)
    
```

```python
athlete.values()
```

```python
# Looping through both keyvalue pair

for keyvalue in athlete.items():
    print(keyvalue)
```

```python
athlete.items()
```


## 10. List Comprehensions


```python
my_friends = ["nathan", "rayan", "aaron", "edwin"]
```

```python
update_my_friends = [friends.title() for friends in my_friends]
```

```python
update_my_friends
```

```python

my_friends_2 = ["nathan ake", "rayan ramsdale", "aaron giggs", "edwin parker"]
```

```python
update_my_friends_2 = [friends.title() for friends in my_friends_2]
```

```python
update_my_friends_2
```

```python
new_update = ["Mr. " + friends for friends in update_my_friends_2]
```

```python
new_update
```


## 12. Positional Arguments


```python

print("Hi there! {} {} ".format( "Brad", "Haddian"))
```

```python

def updating_welcoming(name):
    print("Hi there! {} ".format(name) )
    
```

```python
updating_welcoming("Alex")
```

```python
persons = ["Alan", "David", "Jerry", "Tim"]
```

```python

for person in persons:
    updating_welcoming(person)
    
```


## 13. Multiple Positional Arguments


```python

def updating_info(name, country):
    print("Hi there! {} from {} ".format(name, country))
    
```

```python
updating_info("Jacob", "Italy")
```

```python
updating_info("Albert", "Germany")
```
