## 5. IF, ELSE, Elif Statement

```python
# If Statement

Scientist = "Sir Isaac Newton"

if Scientist == "Sir Isaac Newton":
    print("He is a great scientist")
```

```python

if Scientist != "Newton":
    print("He is not a scientist")
```

```python

my_num = 4 

if my_num <= 5:
    print("The number is less than 5")
    
```

```python
# Else Statement

my_num = 12 

if my_num <= 6:
    print("The number is less than 6")
else:
    print('The number is greater than 6')
    
```

```python
# Elif Statement

my_num = 3

if my_num <= 6:
    print("The number is less than 6")
elif my_num >= 10:
    print("The number is greater than 10")
else:
    print('The number is between 6 and 9')
```

#### If we look for specific condition, we use break after it, and it will not loop back on top over again, as it has been found

```python
# break : It helps you to stop the loop

for countries in country:
    print(countries)
    if countries == "Mexico":
        print("Taco's are best in your country \n")
    elif countries == "Spain":
        print("A great country to visit \n")
        break
    else:
        print("These Countries i didn't get to visit yet")
        print("I will visit this country soon \n")
```

```python
# continue : It is a condition statement block

for countries in country:
    print(countries)
    if countries == "Mexico":
        print("Taco's are best in your country \n")
        continue
    elif countries == "Spain":
        print("A great country to visit \n")
        break
    else:
        print("These Countries i didn't get to visit yet")
        print("I will visit this country soon \n")
```

```python
# pass : This is just a placeholder which does nothing

for countries in country:
    print(countries)
    if countries == "Mexico":
        print("Taco's are best in your country \n")
        continue
    elif countries == "Spain":
        print("A great country to visit \n")
        break
    else:
        pass
        
```


## 8. While Loop


#### A while loop will basically iterate continuously until its condition is no longer true or a break statement is being called.

#### sometimes a while loop basically can become infinite, So it is so important to use the break statement or just turning your condition to false.


```python

balls_bowled = 0            # You have bowled 0 balls
total_balls_bowled = 6      # An over needs 6 balls

while True:                 # Keep playing forever, unless I tell you to stop
    print("Ball Bowled well")     # you bowl one ball and say it was good.
    balls_bowled = balls_bowled + 1     # You increase the number of balls by 1.
     ## or if balls_bowled += 1  
                  
    if balls_bowled == total_balls_bowled:   # Have we bowled 6 balls yet?
        print("The over is done")           # If YES: Say the over is done
        break           # Stop the loop
        
```
