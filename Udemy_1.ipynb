
## 2. Different Data Types

```python
#This is a Strings data type

String_one = "This is my String"
print(String_one)
type(String_one)
```

```python
#This is a Integer data type : (Whole Numbers)

Integer_one = 26
print(Integer_one)
type(Integer_one)
```

```python
Integer_one + 20
```

```python
#This is a Float data type: (Decimal Numbers)

Float_one = 26.01
print(Float_one)
type(Float_one)
```

```python
Float_one - 10.02
```

```python
#This is a Boolean data type     # True/False or 1/0

Boolean_one = True  # or can be False
print(Boolean_one)
type(Boolean_one)
```

```python
## inside a " " 
Boolean_two = "True" 
print(Boolean_two)
type(Boolean_two)
```

```python
#Boolean data type 1/0

Boolean_1 = bool(0)
print(Boolean_1)
type(Boolean_1)
```

```python
#This is a List data type : list[] is mutable - update the values 

A_list = [ 12, 16, 26.01, True, 36, "Doll" ]    #Index Numbering 
        #  0,   1,    2,   3,   4,    5 
```

```python
A_list
```

```python
type(A_list) 
```

```python
A_list[1]
```

```python
A_list[1] = 15
```

```python
A_list
```

```python
#This is a Tuple data type : Tuple() is Immutable - You cannot update the values 

A_tuple = (12, 17, 22, 26, 36, 60) 
```

```python
type(A_tuple)
```

```python
A_tuple[1]
```

```python
#This is a Set data type : Set{} is Immutable - You cannot update the values  
                                # removes duplicated values

A_Set = { 12, 15, 16, 22, 24, 24, 26} 
       
```

```python
type(A_Set)
```

```python
print(A_Set)
```

```python
## To remove duplicated values we can use set 

B_list = [12, 12, 16, 26.01, True, True, 36, "Doll"]
```

```python
set(B_list)
```

```python
## To return to list

list(set(B_list))
```

```python
type(list(set(B_list)))
```
